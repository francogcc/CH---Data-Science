```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from lxml import etree
from datetime import date
from datetime import datetime
import seaborn as sns
from lxml import etree
```

```python
url = requests.get('https://listado.mercadolibre.com.pe/monitor')
url.status_code
```

```python
url.content
```

```python
soup = BeautifulSoup(url.content, 'html.parser')
```

```python
titulos = soup.find_all('h2', attrs={"class":'ui-search-item__title shops__item-title'})
```

```python
len(titulos)
```

```python
print(titulos[0].dtypes)
```

```python
titulos[0].text
```

```python
for i in titulos:
    print(i.text)
```

```python
dom = etree.HTML(str(soup))
precios = dom.xpath('/html/body/main/div/div[2]/section/ol/li[1]/div/div/div[2]/div[2]/div[1]/div[1]/div/div/div/span[1]/span[2]/span[2]')
precios[0]
```

```python
# Descubrimos un patron en las consultas 

# '/html/body/main/div/div[2]/section/ol/li[1]/div/div/div[2]/div[2]/div[1]/div[1]/div/div/div/span[1]/span[2]/span[2]'
# '/html/body/main/div/div[2]/section/ol/li[2]/div/div/div[2]/div[2]/div[1]/div[1]/div/div/div/span[1]/span[2]/span[2]'
# '/html/body/main/div/div[2]/section/ol/li[3]/div/div/div[2]/div[3]/div[1]/div[1]/div/div/div/span[1]/span[2]/span[2]'
# '/html/body/main/div/div[2]/section/ol/li[4]/div/div/div[2]/div[2]/div[1]/div[1]/div/div/div/span[1]/span[2]/span[2]'
```

```python
precios_automaticos = []
for i in range(1, len(titulos)+1):
    x = dom.xpath(f'/html/body/main/div/div[2]/section/ol/li[{i}]/div/div/div[2]/div[2]/div[1]/div[1]/div/div/div/span[1]/span[2]/span[2]')
    try:
        precios_automaticos.append(x[0].text)

    except:
        precios_automaticos.append('Error')
```

```python
df = pd.DataFrame({'Monitor':titulos, 'Precio':precios_automaticos})
df.dtypes
```

```python
# Buscamos si algunos datos no siguen el patron utilizado en el for:
df[df['Precio']=='Error']
```

```python
df 
```

```python
# Buscamos subsanar ese unico elemento que no sigue dicho patron:
# Hallamos cual es el indice del Precio a reemplazar
reemplazar = df[df['Precio']=='Error'].index[0]
```

```python
# Dado que el contador empieza en 1 y no 0, aadimos una unidad a nuestro contador original
x = dom.xpath(f'/html/body/main/div/div[2]/section/ol/li[{reemplazar+1}]/div/div/div[2]/div[3]/div[1]/div[1]/div/div/div/span[1]/span[2]/span[2]')
precios_automaticos[reemplazar] = x[0].text
precios_automaticos[reemplazar]
```

```python
df = pd.DataFrame({'Monitor':titulos, 'Precio':precios_automaticos})
df.dtypes
```
